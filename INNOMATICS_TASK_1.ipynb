```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters for the Poisson distribution
lambda_orders_per_hour = 10  # Average number of incoming orders per hour
hours_in_day = 24  # Simulate for 24 hours

# Generate random order counts for each hour using Poisson distribution
order_counts = poisson.rvs(mu=lambda_orders_per_hour, size=hours_in_day)

# Plotting the incoming orders over the 24-hour period
plt.figure(figsize=(10, 6))
plt.plot(range(hours_in_day), order_counts, marker='o', linestyle='-', color='b')
plt.title("Simulated Incoming Orders Per Hour (Poisson Distribution)")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.xticks(range(0, 24, 1))
plt.show()

# Summary of the orders
order_counts
```
